```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from datetime import datetime
```

a) Look for the missing values in all the columns and either impute them (replace with mean,
median, or mode) or drop them. Justify your action for this task.

```python
missing_values = car_data.isnull().sum()
print(missing_values) # Checking for missing values

missing_columns = []
for col, value in missing_values.items():
    if value > 0:
        missing_columns.append(col)

print("Columns with missing values:", missing_columns)
```

```python
for col in missing_columns:
    if car_data[col].dtype == 'object':
        car_data[col].fillna(car_data[col].mode()[0], inplace=True)
    else:
        car_data[col].fillna(car_data[col].mean(), inplace=True)
```

b) Remove the units from some of the attributes and only keep the numerical values (for
example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from
“New_price”).

```python
column_data_typ = car_data.dtypes

# Print the data types of columns
print(column_data_typ)
```

```python
car_data
```

```python
car_data['Mileage'] = car_data['Mileage'].astype(str)
car_data['Engine'] = car_data['Engine'].astype(str)
car_data['Power'] = car_data['Power'].astype(str)
car_data['New_Price'] = car_data['New_Price'].astype(str)

car_data['Mileage'] = car_data['Mileage'].str.extract('(\d+.\d+)').astype(float)


car_data['Engine'] = car_data['Engine'].str.replace(' CC', '')
car_data['Engine'] = pd.to_numeric(car_data['Engine'], errors='coerce')


default_engine_value = 0
car_data['Engine'] = car_data['Engine'].fillna(default_engine_value).astype(int)

car_data['Power'] = car_data['Power'].str.extract('(\d+.\d+)').astype(float)
car_data['New_Price'] = car_data['New_Price'].str.extract('(\d+.\d+)').astype(float)
```

```python
car_data
```

C) Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot
encoded value.

```python
label_encoder = LabelEncoder()
car_data['Fuel_Type_Label'] = label_encoder.fit_transform(car_data['Fuel_Type'])
car_data['Transmission_Label'] = label_encoder.fit_transform(car_data['Transmission'])

onehot_encoder = OneHotEncoder(sparse=False)

fuel_type_encoded = onehot_encoder.fit_transform(car_data[['Fuel_Type_Label']])
transmission_encoded = onehot_encoder.fit_transform(car_data[['Transmission_Label']])

fuel_type_encoded_df = pd.DataFrame(fuel_type_encoded, columns=['Fuel_Type_' + str(i) for i in range(fuel_type_encoded.shape[1])])
transmission_encoded_df = pd.DataFrame(transmission_encoded, columns=['Transmission_' + str(i) for i in range(transmission_encoded.shape[1])])

car_data = pd.concat([car_data, fuel_type_encoded_df, transmission_encoded_df], axis=1)

car_data.drop(['Fuel_Type', 'Transmission', 'Fuel_Type_Label', 'Transmission_Label'], axis=1, inplace=True)

```
